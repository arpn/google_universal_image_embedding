```python
import pandas as pd
import sh
import os
import torch
import timm
from torchvision.transforms.functional import resize, normalize, pil_to_tensor
from datasets import load_dataset
from tqdm.notebook import tqdm
```

Copy top-100 landmarks to `train/` and `val/` with a 90-10 split.

```python
top_classes = 200
max_examples_per_class = 500
landmarks = pd.read_csv('../input/landmark-recognition-2021/train.csv')
print('Total number of examples:', len(landmarks))
print('Total number of classes:', landmarks.landmark_id.nunique())
top_landmarks = landmarks.groupby('landmark_id').count().sort_values('id', ascending=False).head(top_classes).index
landmark_id_to_class = {id: f'landmark_{cls}' for cls, id in enumerate(top_landmarks)}
for landmark_id in tqdm(top_landmarks):
    examples = landmarks.loc[landmarks.landmark_id == landmark_id].id.head(max_examples_per_class)
    cls = landmark_id_to_class[landmark_id]
    sh.mkdir('-p', f'train/{cls}')
    sh.mkdir('-p', f'val/{cls}')
    for i, example in enumerate(tqdm(examples)):
        # Choose split
        if i / len(examples) < 0.9:
            split = 'train'
        else:
            split = 'val'
        src = f'../input/landmark-recognition-2021/train/{example[0]}/{example[1]}/{example[2]}/{example}.jpg'
        tgt = f'{split}/{cls}/{example}.jpg'
        if not os.path.isfile(tgt):
            sh.cp(src, tgt)
```
